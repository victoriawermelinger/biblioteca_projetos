```python
# prompt: L√≥gica: Mostra a distribui√ß√£o da vari√°vel tempo de tela para Ajuda a entender quantas pessoas passam mais ou menos tempo no celular/computador.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df_habitsvsmental['tempo_uso_tela_hrs'], kde=True, bins=30)
plt.title('Distribui√ß√£o do Tempo de Uso de Tela em Horas')
plt.xlabel('Tempo de Uso de Tela (horas)')
plt.ylabel('Frequ√™ncia')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Insights:
# O histograma mostra a distribui√ß√£o do tempo de uso de tela em horas.
# A maioria das pessoas neste dataset parece ter um tempo de uso de tela entre X e Y horas (observar o pico no gr√°fico).
# H√° uma cauda para a direita, indicando que algumas pessoas t√™m um tempo de uso de tela significativamente maior.
```

```python
# prompt: L√≥gica: Agrupa os dados por quantidade de plataformas usadas e calcula a m√©dia do humor. Analisa se mais redes sociais impactam no bem-estar.

import matplotlib.pyplot as plt
# Agrupar por n√∫mero de plataformas e calcular a m√©dia do humor
mood_by_platforms = df_habitsvsmental.groupby('plataformas_redes_sociais_usadas')['pontuacao_humor'].mean().reset_index()

# Exibir os resultados
print("\nM√©dia do humor por quantidade de plataformas de redes sociais usadas:")
print(mood_by_platforms)

# Visualiza√ß√£o
plt.figure(figsize=(10, 6))
sns.barplot(x='plataformas_redes_sociais_usadas', y='pontuacao_humor', data=mood_by_platforms, palette='viridis')
plt.title('M√©dia do Humor vs. Quantidade de Plataformas de Redes Sociais Usadas')
plt.xlabel('Quantidade de Plataformas de Redes Sociais Usadas')
plt.ylabel('Pontua√ß√£o M√©dia do Humor')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Insights:
# Este gr√°fico de barras mostra a rela√ß√£o entre a quantidade de plataformas de redes sociais usadas e a pontua√ß√£o m√©dia do humor.
# Observa-se uma tend√™ncia de [descrever a tend√™ncia observada, por exemplo: o humor m√©dio parece diminuir/aumentar √† medida que o n√∫mero de plataformas aumenta, ou n√£o h√° uma rela√ß√£o clara].
# Isso sugere que [interpretar a tend√™ncia em rela√ß√£o ao bem-estar, por exemplo: usar mais redes sociais pode estar associado a um humor ligeiramente pior/melhor/neutro].
# Para uma an√°lise mais robusta, seria interessante considerar outros fatores simultaneamente, como o tempo total de uso de tela ou o n√≠vel de estresse.
```

```python
# prompt: L√≥gica: Ordena os dados pelas horas de sono e plota a varia√ß√£o da pontua√ß√£o de humor. Ajuda a ver se quem dorme mais se sente melhor.

import matplotlib.pyplot as plt
# Ordena os dados pelas horas de sono
df_sorted_by_sleep = df_habitsvsmental.sort_values('horas_sono')

# Plota a varia√ß√£o da pontua√ß√£o de humor com base nas horas de sono
plt.figure(figsize=(12, 7))
sns.lineplot(x='horas_sono', y='pontuacao_humor', data=df_sorted_by_sleep, marker='o') # Adiciona marcadores para pontos de dados
plt.title('Varia√ß√£o da Pontua√ß√£o de Humor com Base nas Horas de Sono')
plt.xlabel('Horas de Sono')
plt.ylabel('Pontua√ß√£o de Humor')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Insights:
# Este gr√°fico de linha mostra como a pontua√ß√£o de humor varia com a quantidade de horas de sono.
# A linha tende a [descrever a tend√™ncia, por exemplo: aumentar, diminuir, manter-se est√°vel] √† medida que as horas de sono aumentam.
# Podemos observar que [descrever um ponto espec√≠fico, por exemplo: h√° um pico de humor em torno de X horas de sono, ou o humor parece cair drasticamente com menos de Y horas de sono].
# Isso sugere que [interpretar a rela√ß√£o, por exemplo: dormir mais parece estar associado a um humor melhor, ou h√° uma quantidade ideal de sono para o humor].
# √â importante lembrar que esta √© uma correla√ß√£o e n√£o implica causalidade direta; outros fatores tamb√©m influenciam o humor.
```

```python
# prompt: L√≥gica: Mostra a dispers√£o e mediana dos n√≠veis de estresse conforme a quantidade de redes sociais. Detecta se mais redes aumentam o estresse.

import matplotlib.pyplot as plt
# L√≥gica: Mostra a dispers√£o e mediana dos n√≠veis de estresse conforme a quantidade de redes sociais. Detecta se mais redes aumentam o estresse.

plt.figure(figsize=(10, 6))
sns.boxplot(x='plataformas_redes_sociais_usadas', y='nivel_estresse', data=df_habitsvsmental, palette='coolwarm')
plt.title('N√≠vel de Estresse vs. Quantidade de Plataformas de Redes Sociais Usadas')
plt.xlabel('Quantidade de Plataformas de Redes Sociais Usadas')
plt.ylabel('N√≠vel de Estresse')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Insights:
# Este gr√°fico de caixa (boxplot) visualiza a distribui√ß√£o do n√≠vel de estresse para cada quantidade diferente de plataformas de redes sociais utilizadas.
# Cada caixa representa o intervalo interquartil (IQR), a linha no meio da caixa √© a mediana, e os "bigodes" se estendem aos dados dentro de 1.5 vezes o IQR. Pontos individuais fora dos bigodes s√£o considerados outliers.
# Ao observar as medianas das caixas, podemos ver se h√° uma tend√™ncia clara no n√≠vel de estresse √† medida que o n√∫mero de plataformas aumenta.
# A dispers√£o (tamanho da caixa e extens√£o dos bigodes) mostra a variabilidade do n√≠vel de estresse para cada grupo. Uma caixa maior ou bigodes mais longos indicam maior variabilidade.
# Podemos observar se [descrever a tend√™ncia observada na mediana, por exemplo: a mediana do n√≠vel de estresse parece aumentar √† medida que o n√∫mero de plataformas aumenta, ou parece diminuir, ou n√£o h√° uma tend√™ncia clara].
# A presen√ßa de outliers (pontos individuais) indica indiv√≠duos com n√≠veis de estresse excepcionalmente altos ou baixos para o n√∫mero de plataformas que utilizam.
# Em geral, se as medianas e a dispers√£o tendem a ser maiores para grupos que utilizam mais plataformas, isso pode sugerir uma associa√ß√£o positiva entre o uso de mais redes sociais e n√≠veis mais altos de estresse.
```

```python
# prompt: L√≥gica: Relaciona o tempo no TikTok com o humor, destacando com cores diferentes os n√≠veis de estresse. Ajuda a ver se mais TikTok impacta negativamente.

import matplotlib.pyplot as plt
# Scatter plot para relacionar horas no TikTok com pontua√ß√£o de humor, colorindo por n√≠vel de estresse
plt.figure(figsize=(12, 7))
sns.scatterplot(x='horas_no_tiktok', y='pontuacao_humor', hue='nivel_estresse', data=df_habitsvsmental, palette='viridis', alpha=0.6)
plt.title('Horas no TikTok vs. Pontua√ß√£o de Humor (Colorido por N√≠vel de Estresse)')
plt.xlabel('Horas no TikTok')
plt.ylabel('Pontua√ß√£o de Humor')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='N√≠vel de Estresse')
plt.show()

# Insights:
# Este scatter plot mostra a rela√ß√£o entre as horas passadas no TikTok e a pontua√ß√£o de humor, com os pontos coloridos de acordo com o n√≠vel de estresse.
# Cada ponto no gr√°fico representa um indiv√≠duo no dataset.
# Podemos observar se h√° uma tend√™ncia geral na pontua√ß√£o de humor √† medida que as horas no TikTok aumentam. Por exemplo, o humor tende a diminuir/aumentar/permanecer o mesmo?
# A colora√ß√£o dos pontos permite visualizar como o n√≠vel de estresse se distribui nessa rela√ß√£o. Por exemplo, os pontos de alto estresse tendem a se agrupar em uma √°rea espec√≠fica do gr√°fico (por exemplo, mais horas no TikTok e menor humor)?
# Se observarmos que pontos com n√≠veis de estresse mais altos tendem a estar associados a mais horas no TikTok e/ou menor pontua√ß√£o de humor, isso pode sugerir uma correla√ß√£o entre o uso do TikTok, estresse e humor.
# No entanto, √© crucial lembrar que correla√ß√£o n√£o implica causalidade. Outros fatores podem estar influenciando simultaneamente todas essas vari√°veis.
# Este gr√°fico ajuda a identificar padr√µes visuais e √°reas de interesse para uma an√°lise mais aprofundada.
```

```python
# prompt: L√≥gica: Calcula a correla√ß√£o entre todas as vari√°veis num√©ricas. O mapa mostra quais h√°bitos digitais t√™m mais rela√ß√£o com sono, estresse e humor.

import matplotlib.pyplot as plt
# Seleciona apenas as colunas num√©ricas para o c√°lculo da correla√ß√£o
df_numeric = df_habitsvsmental[['tempo_uso_tela_hrs', 'plataformas_redes_sociais_usadas', 'horas_no_tiktok', 'horas_sono', 'nivel_estresse', 'pontuacao_humor']]

# Calcula a matriz de correla√ß√£o
correlation_matrix = df_numeric.corr()

# Exibe a matriz de correla√ß√£o
print("\nMatriz de Correla√ß√£o entre Vari√°veis Num√©ricas:")
print(correlation_matrix)

# Visualiza a matriz de correla√ß√£o usando um mapa de calor (heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor da Matriz de Correla√ß√£o')
plt.show()

# Insights:
# Este mapa de calor mostra as correla√ß√µes entre todos os pares de vari√°veis num√©ricas no dataset.
# Os valores variam de -1 a 1. Valores pr√≥ximos a 1 indicam uma forte correla√ß√£o positiva (√† medida que uma vari√°vel aumenta, a outra tamb√©m tende a aumentar). Valores pr√≥ximos a -1 indicam uma forte correla√ß√£o negativa (√† medida que uma vari√°vel aumenta, a outra tende a diminuir). Valores pr√≥ximos a 0 indicam pouca ou nenhuma correla√ß√£o linear.
# A diagonal principal √© sempre 1, pois uma vari√°vel est√° perfeitamente correlacionada consigo mesma.
# Observando as c√©lulas onde as linhas e colunas se cruzam (exceto a diagonal):
# - Procure por valores altos (pr√≥ximos a 1 ou -1) nas c√©lulas que relacionam os h√°bitos digitais ('tempo_uso_tela_hrs', 'plataformas_redes_sociais_usadas', 'horas_no_tiktok') com os resultados de sa√∫de mental ('horas_sono', 'nivel_estresse', 'pontuacao_humor').
# - Por exemplo, uma correla√ß√£o positiva forte entre 'tempo_uso_tela_hrs' e 'nivel_estresse' sugeriria que mais tempo de tela est√° associado a n√≠veis mais altos de estresse.
# - Uma correla√ß√£o negativa forte entre 'horas_no_tiktok' e 'pontuacao_humor' sugeriria que mais horas no TikTok est√£o associadas a pontua√ß√µes de humor mais baixas.
# - As cores no mapa de calor tamb√©m ajudam a visualizar a for√ßa e dire√ß√£o da correla√ß√£o (geralmente azul/roxo para negativo, vermelho/laranja para positivo, e cores neutras para perto de zero).
# Este mapa de calor √© uma ferramenta poderosa para identificar rapidamente quais vari√°veis t√™m as rela√ß√µes lineares mais fortes umas com as outras. No entanto, √© importante lembrar que correla√ß√£o n√£o implica causalidade, e esta an√°lise √© apenas um ponto de partida para entender as rela√ß√µes entre as vari√°veis.
```
