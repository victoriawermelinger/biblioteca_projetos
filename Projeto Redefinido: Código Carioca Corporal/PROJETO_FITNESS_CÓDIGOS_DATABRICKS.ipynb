```python
# Importa a biblioteca pandas para manipulação de dados em tabelas (DataFrames)
import pandas as pd
import numpy as np
import pyspark.sql.functions as F
from pyspark.sql.functions import col, sum as spark_sum
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, lpad
```

```python
# Cria um DataFrame pandas a partir do arquivo CSV
df_fitness_23_06_pd = pd.read_csv('/Workspace/Users/wermelinger.v.dg@gmail.com/PROJETO.FITNESS (23.06.2025) -  Código Carioca Corporal.csv')

# Converte o DataFrame pandas para DataFrame Spark
df_fitness_23_06 = spark.createDataFrame(df_fitness_23_06_pd)

# Exibe o DataFrame Spark
display(df_fitness_23_06)
```

```python
# Renomeia as colunas do DataFrame para nomes mais claros e padronizados
novos_nomes_colunas = {
    'Carimbo de data/hora': 'Codigo_ID',
    'Data (00/00/0000)': 'data',
    'peso (00.0)': 'peso',
    'IMC (Indice de Massa Corporal) (00.0)': 'imc',
    'Taxa de Gordura Corporal (00.0) %': 'taxa_gordura_corporal',
    'Kg Massa Gorda (00.0)': 'kg_massa_gorda',
    'Kg Massa Livre de Gordura (00.0)': 'kg_massa_livre_gordura',
    'Massa Muscular (00.0)': 'massa_muscular',
    'Taxa Massa Muscular (00.0)%': 'taxa_massa_muscular',
    'Massa Muscular Esqueletica (00.0)%': 'massa_muscular_esqueletica',
    'Gordura Subcutânea (00.0)%': 'gordura_subcutanea',
    'Gordura Visceral (00.0)': 'gordura_visceral'
}

# Renomeia as colunas do DataFrame Spark conforme o dicionário acima
for nome_original, nome_novo in novos_nomes_colunas.items():
    df_fitness_23_06 = df_fitness_23_06.withColumnRenamed(nome_original, nome_novo)

# Exibe o DataFrame Spark com as colunas renomeadas
display(df_fitness_23_06)
```

```python
# Cria uma especificação de janela para ordenar pela coluna 'data'
window_spec = Window.orderBy("data")

# Adiciona uma coluna 'Codigo_ID' com números sequenciais formatados como "001", "002", etc.
df_fitness_23_06 = df_fitness_23_06.withColumn("Codigo_ID", lpad(row_number().over(window_spec), 3, "0"))

# Exibe o DataFrame Spark com a nova coluna 'Codigo_ID'
display(df_fitness_23_06)
```

```python
# Verifica se há dados nulos em cada coluna
nulos_por_coluna = df_fitness_23_06.select([spark_sum(col(c).isNull().cast("int")).alias(c) for c in df_fitness_23_06.columns])

display(nulos_por_coluna)
```

```python
# Converte a coluna 'data' de string para tipo data no formato 'dd/MM/yyyy'
df_fitness_23_06 = df_fitness_23_06.withColumn("data", to_date("data", "dd/MM/yyyy"))
```
