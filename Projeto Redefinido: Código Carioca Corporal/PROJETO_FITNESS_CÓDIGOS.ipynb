```python
import pandas as pd

# Substitua 'seu_arquivo.csv' pelo caminho do seu arquivo CSV no Google Drive
# Certifique-se de que o Colab está conectado ao Google Drive e que você
# autorizou o acesso.
try:
    df_projeto = pd.read_csv('/home/wermelinger_v/Documentos/GitHub/Codigo_Fonte_WermelingerV/Projeto Redefinido: Código Carioca Corporal/PROJETO.FITNESS (23.06.2025) -  Código Carioca Corporal.csv')

    # Agora você pode trabalhar com o DataFrame 'df_projeto'
    # Por exemplo, para exibir as primeiras 5 linhas:
    print(df_projeto.head())

except FileNotFoundError:
    print("Erro: O arquivo não foi encontrado. Verifique o caminho do arquivo.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo: {e}")
```

```python
# Renomeando as colunas do DataFrame para nomes mais simples
df_projeto.columns = [
    'Código_ID',
    'Data',
    'Peso',
    'IMC',
    'Taxa_Gordura_Corporal',
    'Kg_Massa_Gorda',
    'Kg_Massa_Livre_Gordura',
    'Massa_Muscular',
    'Taxa_Massa_Muscular',
    'Massa_Muscular_Esqueletica',
    'Gordura_Subcutanea',
    'Gordura_Visceral'
]

# Exibindo as primeiras linhas para conferir
print(df_projeto.head())
```

```python
df_projeto['Código_ID'] = df_projeto.index + 1
df_projeto['Código_ID'] = df_projeto['Código_ID'].astype(str).str.zfill(3)
print(df_projeto[['Código_ID']].head())
```

```python
df_projeto
```

```python
import matplotlib.pyplot as plt

# Converter a coluna 'Data' para datetime
df_projeto['Data'] = pd.to_datetime(df_projeto['Data'], format='%d/%m/%Y')

# Agrupar por mês e calcular a média do IMC e da Massa Gorda
df_mensal = df_projeto.resample('M', on='Data').mean(numeric_only=True)


# Plotar a evolução do IMC e da Massa Gorda ao longo dos meses
plt.figure(figsize=(12, 6))
plt.plot(df_mensal.index, df_mensal['IMC'], marker='o', label='IMC')
plt.plot(df_mensal.index, df_mensal['Kg_Massa_Gorda'], marker='s', label='Massa Gorda (kg)')
plt.title('Evolução do IMC e Massa Gorda ao longo dos meses')
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotar a evolução da Massa Muscular e Massa Gorda ao longo dos meses
df_mensal['Massa_Muscular'] = df_mensal['Massa_Muscular'].fillna(0)
df_mensal['Kg_Massa_Gorda'] = df_mensal['Kg_Massa_Gorda'].fillna(0)
df_mensal = df_mensal[['Massa_Muscular', 'Kg_Massa_Gorda']] # Selecionando apenas as colunas relevantes
plt.figure(figsize=(12, 6))
plt.plot(df_mensal.index, df_mensal['Massa_Muscular'], marker='o', label='Massa Muscular (kg)')
plt.plot(df_mensal.index, df_mensal['Kg_Massa_Gorda'], marker='s', label='Massa Gorda (kg)')
plt.title('Evolução de Massa Muscular e Massa Gorda ao longo dos meses')
plt.xlabel('Mês')
plt.ylabel('Valor (kg)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotar a evolução do Peso ao longo dos meses
df_projeto['Peso'] = df_projeto['Peso'].fillna(0)
df_projeto = df_projeto[['Data', 'Peso']]  # Selecionando apenas as colunas relevantes
df_projeto['Data'] = pd.to_datetime(df_projeto['Data'], format='%Y-%m-%d')

# Agrupar por mês e calcular a média do Peso
df_peso_mensal = df_projeto.resample('M', on='Data').mean(numeric_only=True)

plt.figure(figsize=(12, 6))
plt.plot(df_peso_mensal.index, df_peso_mensal['Peso'], marker='o', color='tab:blue', label='Peso (kg)')
plt.title('Evolução do Peso ao longo dos meses')
plt.xlabel('Mês')
plt.ylabel('Peso (kg)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
